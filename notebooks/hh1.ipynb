     guesses = [5.0]
     errors = [10000]

     def cube (x):
         return x**3

     def derivCube (x):
         return 3*x**2

     tolerance = 0.01

     while (errors[-1] > tolerance):
         errors.append(128 - cube(guesses[-1]))
         guesses.append(errors[-1]/derivCube(guesses[-1]) + guesses[-1])

     (errors,guesses)
